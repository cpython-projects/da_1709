**Task 7.**
Влияет ли размер семьи на шанс выживания?

```python
total = len(df[df.Survived == 1])

for n in sorted(df.FamilySize.unique()):
  tmp = df[(df.FamilySize == n) & (df.Survived == 1)]
  print(n, len(tmp), total, f'{round(len(tmp) / total * 100)}%')
```

**Task 8.** Были ли различия в выживании среди пассажиров с разными титулами (Mr, Mrs, Dr и т.д.)?

```python
df_mr = df[(df.Name.str.contains('Mr. ')) & (df.Survived == 1)]
df_mrs = df[df.Name.str.contains('Mrs. ') & (df.Survived == 1)]
df_dr = df[df.Name.str.contains('Dr. ') & (df.Survived == 1)]

print(len(df_mr), len(df_mrs), len(df_dr))
```
