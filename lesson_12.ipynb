```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/cpython-projects/da_1709/refs/heads/main/lesson_12_data.csv')
df
```

# Step 1

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
df.columns.to_list()
```

```python
df.rename(columns={
    'ID': 'id',
    'Name': 'user_name',
    'Age': 'user_age',
    'Salary ($)': 'salary_usd',
    'Hire Date': 'hire_date',
    'Country': 'country',
    'Weight': 'user_weight',
  }, inplace=True)
df
```

# Step 2

```python
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
df
```

# Step 3

```python
df.info()
```

# Step 4

```python
df.gender.unique()
```

```python
df['gender'] = df.gender.astype('category')
df.info()
```

```python
df.country.unique()
```

```python
df['country'] = df.country.replace({
    'UK': 'United Kingdom',
    'USA': 'United States',
    'U.S.A.': 'United States',
    'canada': 'Canada',
})
```

```python
df
```

```python
df['user_name'] = df.user_name.str.title()
df
```

```python
df.user_age.unique()
```

```python
df['user_age'] = df.user_age.replace({
    'Thirty': '30',
    'Twenty': '20'
})
```

```python
df.user_age.unique()
```

```python
df.info()
```

```python
df['user_age'] = pd.to_numeric(df.user_age, errors='coerce')
df.user_age.unique()
```

```python
df.hire_date.unique()
```

```python
#df['hire_date'] = pd.to_datetime(df.hire_date, format='%Y-%m-%d', errors='coerce')
#df.hire_date.unique()
```

```python
from dateutil.parser import parse

df['hire_date'] = df.hire_date.apply(parse)
df.hire_date.unique()
```

```python
df['hire_date'] = pd.to_datetime(df.hire_date, format='%Y-%m-%d', errors='coerce')
df
```

```python
df.info()
```

```python
def weight_converter(user_weight: str) -> float:
  user_weight = user_weight.lower().strip()

  if 'kg' in user_weight:
    return float(user_weight.replace('kg', ''))

  if 'lbs' in user_weight:
    return float(user_weight.replace('lbs', '')) * 0.453592

  if 'g' in user_weight:
    return float(user_weight.replace('g', '')) * 0.001

  return None
```

```python
df['user_weight_kg'] = df.user_weight.apply(weight_converter)
df
```

```python
df.info()
```

```python
age_median = df.user_age.median()
age_median
```

```python
salary_median = df.salary_usd.median()
salary_median
```

```python
df['user_age'] = df['user_age'].fillna(age_median)
df['salary_usd'] = df['salary_usd'].fillna(salary_median)
df
```
