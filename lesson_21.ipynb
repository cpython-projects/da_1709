```python
import sqlite3
```

```python
DB_NAME = 'school'
```

```python
conn = sqlite3.connect(f'{DB_NAME}.db')
```

```python
type(conn)
```

```python
cursor = conn.cursor()
```

```python
create_students = """
  CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    gender TEXT NOT NULL,
    phone_number TEXT NOT NULL,
    email TEXT
  );
"""
```

```python
cursor.execute(create_students)
```

```python
create_teachers = """
CREATE TABLE IF NOT EXISTS teachers (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL
);
"""
```

```python
cursor.execute(create_teachers)
```

```python
create_courses = """
CREATE TABLE IF NOT EXISTS courses (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL
);
"""
```

```python
cursor.execute(create_courses)
```

```python
create_group = """
CREATE TABLE IF NOT EXISTS groups (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    course_id INTEGER NOT NULL,
    FOREIGN KEY (course_id) REFERENCES courses (id)
);
"""
```

```python
cursor.execute(create_group)
```

```python
students_in_groups = """
CREATE TABLE IF NOT EXISTS students_in_groups (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER NOT NULL,
    group_id INTEGER NOT NULL,
    FOREIGN KEY (student_id) REFERENCES students (id),
    FOREIGN KEY (group_id) REFERENCES groups (id)
);
"""
```

```python
cursor.execute(students_in_groups)
```

```python
teachers_in_groups = """
CREATE TABLE IF NOT EXISTS teachers_in_groups (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    teacher_id INTEGER NOT NULL,
    group_id INTEGER NOT NULL,
    FOREIGN KEY (teacher_id) REFERENCES teachers (id),
    FOREIGN KEY (group_id) REFERENCES groups (id)
);
"""
```

```python
cursor.execute(teachers_in_groups)
```

```python
conn.close()
```
